```python
numeric_cols=list(train_data.select_dtypes(include='float64').columns)
cat_cols=list(train_data.select_dtypes(include='object').columns)
cat_cols=cat_cols
cat_cols
```

```python
imputer=SimpleImputer(strategy='most_frequent')
```

```python
imputer.fit(train_data[numeric_cols])
imputer.fit(test_df[numeric_cols])
```

```python
train_data[numeric_cols]=imputer.transform(train_data[numeric_cols])
test_df[numeric_cols]=imputer.transform(test_df[numeric_cols])
test_data[numeric_cols]=imputer.transform(test_data[numeric_cols])

imputer.fit(train_data[cat_cols])
train_data[cat_cols]=imputer.transform(train_data[cat_cols])
test_df[cat_cols]=imputer.transform(test_df[cat_cols])
test_data[cat_cols]=imputer.transform(test_data[cat_cols])
```

```python
train_data['HomePlanet']=train_data['HomePlanet'].fillna('Unknown')
test_df['HomePlanet']=test_df['HomePlanet'].fillna('Unknown')
test_data['HomePlanet']=test_data['HomePlanet'].fillna('Unknown')
```
